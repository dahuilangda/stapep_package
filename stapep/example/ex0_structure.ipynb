```python
from stapep.structure import Structure

# Initialize a sequence and output directory
# 序列规则:
#   使用氨基酸单字母缩写用
#   单字母氨基酸之间可以连续写，也可以用'-'分隔，非单字母氨基酸之间必须用'-'分隔。例如: Ac-BATPR8RRR-Aib-LBRR3FKRLQ-NH2
#   N端可以使用'Ac-'开头, 添加Acetyl group
#   C端可以使用'-NH2'结尾, 添加Amide group
#   Aib是2-Aminoisobutyric acid
#   B是Nle的缩写，Nle是Norleucine

seq = 'Ac-BATP-R8-RRR-Aib-BLBR-R3-FKRLQ'
st = Structure(verbose=True)
```

```python
from stapep.structure import Structure

# De novo 3D structure prediction using ESMFOLD
st = Structure(verbose=True)
st.de_novo_3d_structure(seq, output_pdb='data/denovo.pdb')
```

```python
from stapep.structure import Structure, AlignStructure

# homology modeling
st = Structure(verbose=True)
st.generate_3d_structure_from_template(seq=seq, 
                                       output_pdb='data/homology_model.pdb', 
                                       template_pdb='data/template.pdb')
```

```python
# 首先，读取两个PDB文件的内容
with open("data/template.pdb", "r") as f:
    template_data = f.read()

with open("data/homology_model.pdb", "r") as f:
    homology_model_data = f.read()

p = py3Dmol.view()
p.addModel(homology_model_data, format="pdb")
p.setStyle({'cartoon': {'color':'spectrum'}})
p.addModel(template_data, format="pdb")
p.setStyle({'model': 1}, {'cartoon': {'color':'gray'}}) 
p.zoomTo()
```

```python
# align structure to template
AlignStructure.align(ref_pdb='data/template.pdb',
                     pdb='data/homology_model.pdb',
                     output_pdb='data/aligned.pdb')
```

```python
with open("data/aligned.pdb", "r") as f:
    aligned_data = f.read()

p = py3Dmol.view()
p.addModel(aligned_data, format="pdb")
p.setStyle({'cartoon': {'color':'spectrum'}})
p.addModel(template_data, format="pdb")
p.setStyle({'model': 1}, {'cartoon': {'color':'gray'}}) 
p.zoomTo()
```
